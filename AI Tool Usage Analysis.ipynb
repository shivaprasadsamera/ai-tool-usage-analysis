### Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset

```python
# Load the dataset
file_path = "Generative AI in Education.csv"
df = pd.read_csv('/content/Dataset/Students.csv')
```

```python
# --- Raw Exploration Section (from your code) ---

print("First 5 rows:")
df.head()
```

### Step 4: Check Unique Values per Column

```python
# Count unique values per column
print("\nUnique value counts per column:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()}")
```

### Step 5: Handle Missing Values

```python
# Checking nulls and filling with mode
print("\nMissing values before cleaning:")
print(df.isnull().sum())
for col in df.columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)
print("\nMissing values after filling:")
df.isnull().sum()
```

### Step 6: Visualizations and Insights

#### 6.1 Distribution by Stream

```python
# Distribution of Streams
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='Stream', palette='Set2')
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width() / 2, height + 0.5, int(height),
                ha='center', va='bottom', fontsize=8, color='black')
plt.title("Distribution of Students by Stream")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 6.2 Most Used AI Tools

```python
# Most Used AI Tools
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, y='AI_Tools_Used', order=df['AI_Tools_Used'].value_counts().index, palette='coolwarm')
for p in ax.patches:
    count = int(p.get_width())
    ax.text(p.get_width() + 1, p.get_y() + p.get_height() / 2,
            str(count), ha='left', va='center', fontsize=8)
plt.title("Most Common AI Tools Used")
plt.tight_layout()
plt.show()
```

#### 6.3 AI Usage vs Impact on Grades

```python
# AI Usage vs. Grades
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='Daily_Usage_Hours', y='Impact_on_Grades',hue='Daily_Usage_Hours', palette='Blues', legend=False)
plt.title("AI Usage vs Grade")
plt.tight_layout()
plt.show()
```

#### 6.4 Trust in AI Tools

```python
# Trust in AI Histogram
plt.figure(figsize=(10, 6))
df['Trust_in_AI_Tools'] = pd.to_numeric(df['Trust_in_AI_Tools'], errors='coerce')
ax = sns.histplot(df['Trust_in_AI_Tools'].dropna(), bins=10, kde=True, color='green')
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width() / 2, height + 0.5, int(height),
                ha='center', va='bottom', fontsize=8, color='black')
plt.title("Trust in AI Distribution", fontsize=14)
plt.xlabel("Trust Level (1-5)")
plt.ylabel("Number of students")
plt.tight_layout()
plt.show()
```

#### 6.5 Average Usage per Stream

```python
# Average AI Usage by Stream
plt.figure(figsize=(10, 5))
avg_usage = df.groupby("Stream")['Daily_Usage_Hours'].mean().sort_values()
avg_usage.plot(kind='barh', color='purple')
for i, v in enumerate(avg_usage):
    plt.text(v, i, f"{v:.2f}", va='center')
plt.title("Average AI Usage per Stream", fontsize=14)
plt.xlabel("Hours per Week")
plt.tight_layout()
plt.show()

```

#### 6.6 Correlation Heatmap

```python
# Correlation Analysis (only numeric columns)
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
